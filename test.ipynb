```python
audio_file = "dataset/audio_test/1/2.wav"

from pyAudioAnalysis import audioBasicIO, ShortTermFeatures
import numpy as np

sampling_rate, signal = audioBasicIO.read_audio_file(audio_file)
print(sampling_rate)
signal = audioBasicIO.stereo_to_mono(signal)
print(signal)
```
